# Filteri

```python
def contrast():
    t=dugme1.get()
    slika_nova=function_c(slika,t)
    nova = Image.fromarray(np.uint8(slika_nova))
    nova=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')
    
    histogram(np.uint8(slika_nova))
    root.mainloop()
    
def function_c(image,t):
    br=128
    nova=np.zeros((512,512,3),dtype=np.uint8)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for k in range(image.shape[2]):
                x=image[i,j,k]
                if(x<br):
                    x=x-t
                    if x<0:
                        x=0
                else:
                    x=x+t
                    if x>255:
                        x=255
                
                nova[i,j,k]=x
    return nova
```

```python
def highlights():
    t=dugme2.get()
    slika_nova=function_h(slika,t)
    nova = Image.fromarray(np.uint8(slika_nova))
    nova=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')
    histogram(np.uint8(slika_nova))
    root.mainloop()
    
def function_h(image,t):
    br=180
    nova=np.zeros((512,512,3),dtype=np.uint8) 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for k in range(image.shape[2]):
                x=image[i,j,k]
                if(x>br):
                    x=x+t
                    if x>255:
                        x=255
                
                nova[i,j,k]=x
    return nova
```

```python
def shadows():
    t=dugme3.get()
    slika_nova=function_s(slika,t)
    nova = Image.fromarray(np.uint8(slika_nova))
    nova=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')
    histogram(np.uint8(slika_nova))
    root.mainloop()
    
def function_s(image,t):
    br=80
    nova=np.zeros((512,512,3),dtype=np.uint8) 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for k in range(image.shape[2]):
                x=image[i,j,k]
                if(x<br):
                    x=x-t
                    if x<0:
                        x=0
                
                nova[i,j,k]=x
    return nova
```

```python
def brightness(): 
    slika_f=prolazak(slika*float(dugme4.get())) 
    cuvanje=slika_f
    nova = Image.fromarray(np.uint8(slika_f)) 
    nova=ImageTk.PhotoImage(nova) 
    w.create_image(0, 0, image=nova, anchor='nw') 
    histogram(np.uint8(cuvanje))
    root.mainloop()
```

```python
def fade():
    t=dugme5.get()
    sl_nule=np.zeros((512,512,3),dtype=np.uint8)
    R=slika.T[0]+float(t)
    G=slika.T[1]+float(t)
    B=slika.T[2]+float(t)
    sl_nule.T[0]=np.array(prolazak1(R),dtype=np.uint8)
    sl_nule.T[1]=np.array(prolazak1(G),dtype=np.uint8)
    sl_nule.T[2]=np.array(prolazak1(B),dtype=np.uint8)
    nova1=Image.fromarray(sl_nule)
    nova1 = ImageTk.PhotoImage(nova1)
    w.create_image(0, 0, image=nova1, anchor='nw')
    histogram(sl_nule)
    root.mainloop()
```

```python
def warmth():
    t=dugme6.get()
    R=slika.T[0]
    G=slika.T[1]
    B=slika.T[2]
    if(t>0):
        R=prolazak1((1+t)*R) 
        G=prolazak1((1-t)*G)
        B=prolazak1((1-t)*B)
    else:
        R=prolazak1((1+t)*R)
        G=prolazak1((1-t)*G)
        B=prolazak1((1-t)*B)   
    nule=np.zeros((512,512,3),dtype=np.uint8)
    nule.T[0]=R
    nule.T[1]=G
    nule.T[2]=B
    nova=Image.fromarray(nule)
    nova=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')
    histogram(nule)
    root.mainloop()
```

```python
def saturation(): 
    image=Image.fromarray(slika).convert('HSV')
    H=image.getchannel('H')
    V=image.getchannel('V')
    
    slika_niz=np.array(image)
    slika_niz=dugme7.get()*slika_niz
    saturation=prolazak1(slika_niz.T[1].T) 
    satur_new = Image.fromarray(np.uint8(saturation))

    HSV = Image.merge('HSV', (H, satur_new, V))
    R,G,B = HSV.convert('RGB').split() 
    nova = Image.merge('RGB',(R,G,B))
    nova1=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova1, anchor='nw')
    histogram(np.array(nova))
    root.mainloop()
```

```python
def sharpness():
    image=Image.fromarray(slika).convert('HSV')
    H=image.getchannel('H')
    S=image.getchannel('S')
    V=image.getchannel('V') 

    h=np.ones((11,11))/121
    V=signal.convolve2d(V,h,'same') 

    newV=np.array(prolazak1(V),dtype=np.uint8)
    newV=Image.fromarray(newV)
    HSV = Image.merge('HSV', (H, S, newV))
    R,G,B = HSV.convert('RGB').split()
    nova = Image.merge('RGB',(R,G,B))
    nova=np.array(nova)

    c=float(dugme8.get())
    unsharp=(c/(2*c-1))*slika-((1-c)/(2*c-1))*nova
    unsharp=np.array(prolazak(unsharp),dtype=np.uint8)
    nova=Image.fromarray(unsharp)
    nova = ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')
    histogram(unsharp)
    root.mainloop()
```

```python
def tilt_shift():
    image=Image.fromarray(slika).convert('HSV')
    H=image.getchannel('H')
    S=image.getchannel('S')
    V=image.getchannel('V') 
    
    V_niz=np.array(V)

    h=np.ones((11,11))/121
    Vc=signal.convolve2d(V,h,'same')
    newV=np.array(prolazak1(Vc),dtype=np.uint8)

    for i in range(dugme9.get()):
        V_niz[i]=newV[i]
        V_niz[511-i]=newV[511-i]

    V_niz=Image.fromarray(V_niz)
    HSV = Image.merge('HSV', (H, S, V_niz))
    R,G,B = HSV.convert('RGB').split()
    nova1 = Image.merge('RGB',(R,G,B))
    nova = ImageTk.PhotoImage(nova1)
    w.create_image(0, 0, image=nova, anchor='nw')
    histogram(np.array(nova1))
    root.mainloop()
```

```python
def zoom(): 
    broj = dugme10.get()
    h = int(broj * slika.shape[0])
    g = int(broj * slika.shape[1])

    zoomed = np.zeros(shape=(h , g , 3) , dtype = 'uint8')

    for i in range(0,h):
        for j in range(0,g):
            zoomed[i,j] = slika[int(i/broj) , int(j/broj)]

    cen = int(h // 2)
    h2 = int(cen - 256) 
    w2 = int(cen + 256)        
    zoomed = zoomed[h2:w2 , h2:w2]
    
    zum = Image.fromarray(np.uint8(zoomed))
    zum=ImageTk.PhotoImage(zum)

    w.create_image(0, 0, image=zum, anchor='nw')
    histogram(np.uint8(zoomed))
    root.mainloop()
```

```python
def rotation():
    degree=dugme11.get()
    rads = math.radians(degree)
    image=slika
    rot_img = np.uint8(np.zeros(image.shape))
    
    height = rot_img.shape[0]
    width  = rot_img.shape[1]

    midx,midy = (width//2, height//2)

    for i in range(rot_img.shape[0]):
        for j in range(rot_img.shape[1]):
            x= (i-midx)*math.cos(rads)+(j-midy)*math.sin(rads)
            y= -(i-midx)*math.sin(rads)+(j-midy)*math.cos(rads)

            x=round(x)+midx 
            y=round(y)+midy 

            if (x>=0 and y>=0 and x<image.shape[0] and  y<image.shape[1]):
                rot_img[i,j,:] = image[x,y,:]
                
    nova = Image.fromarray(np.uint8(rot_img ))
    nova=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')       
    histogram(rot_img)
    root.mainloop()
```

```python
def vignette():
    red, kolona = slika.shape[0], slika.shape[1]
    a= np.ones(red)
    b= np.ones(kolona)
    slider = dugme12.get()
    ogranicenje_a = (slider/100) *(1/4) * red 
    ogranicenje_a = int(ogranicenje_a)
    ogranicenje_b=(slider/100) * (1/4) * kolona
    ogranicenje_b = int(ogranicenje_b)
    for i in range(ogranicenje_a):
        a[i] = i / ogranicenje_a
        a[red - 1 - i] = a[i] 

    for i in range(ogranicenje_b):
        b[i] = i / ogranicenje_b
        b[kolona - 1 - i] = b[i]

    maska=np.zeros((red,kolona))
    for i in range(red):
        for j in range(kolona):
            maska[i,j]=a[i]*b[j] 

    output = np.copy(slika)
    for i in range(3):
            output[:,:,i] = output[:,:,i] * maska

    output=output.astype(np.uint8)
    nova = Image.fromarray(output)
    nova=ImageTk.PhotoImage(nova)
    w.create_image(0, 0, image=nova, anchor='nw')
    histogram(output)
    root.mainloop()
```


### Pomocne funkcije

```python
def prolazak(sl):
    for i in range(sl.shape[0]):
        for j in range(sl.shape[1]):
            for k in range(sl.shape[2]):
                if(sl[i,j,k]>255):
                    sl[i,j,k]=255
                if(sl[i,j,k]<0):
                    sl[i,j,k]=0
    return sl
```

```python
def prolazak1(sl):
    for i in range(sl.shape[0]):
        for j in range(sl.shape[1]):
            if(sl[i,j]>255):
                sl[i,j]=255
            if(sl[i,j]<0):
                sl[i,j]=0
    return sl
```

### Refresh

```python
def refresh():
    sl = Image.fromarray(np.uint8(slika))
    sl = ImageTk.PhotoImage(sl)
    w.create_image(0, 0, image=sl, anchor='nw')
    histogram(slika)
    root.mainloop()
```

# Histogram

```python
def histogram(slika):
    plt.ioff()
    fig=plt.figure()
    image=Image.fromarray(np.uint8(slika))
    hst= image.histogram()
    Red=hst[0:256]
    Green=hst[256:512]
    Blue=hst[512:768]
    for i in range(0,256):
        plt.bar(i,Red[i], color='red', alpha=0.7)
        
    for i in range(0,256):
        plt.bar(i,Green[i], color='green', alpha=0.7)
        
    for i in range(0,256):
        plt.bar(i,Blue[i], color='blue', alpha=0.7)
    
    plt.savefig('histogram.jpg')
    
    hist=plt.imread("histogram.jpg")
    hist=ImageTk.PhotoImage(Image.fromarray(hist))
    w.create_image(790, 215, image=hist, anchor='nw')
    root.mainloop()
```
